## Problem 1: Trapezoidal Rule

Write a program to calculate the approximate value of the integral ∫a to b (x^6−5x^5+3)dx using Trapezoidal rule. Divide the integration interval into N number of segments of equal width. The values of a, b and N will be specified as input.
				
The input will consist of three lines, with the first and second lines specifying a and b respectively. The third line will contain the value for N.
				
Round the answer to two decimal places using the Python function round() as below:
ans = round(ans, 2)

Note: You should not use "scipy.integrate" package in your code.
				
Example: 
INPUT: 
3
5
1000
			
OUTPUT:
-1559.05

```python
import numpy as np

a = float(input())
b = float(input())
N = int(input())

r = np.linspace(a,b,N+1)
diff = r[1]-r[0]

ans = 0.0
for i in range(len(r)-1):
    ans+= (diff/2)*((r[i+1]**6 - 5*r[i+1]**5 + 3) + (r[i]**6 - 5*r[i]**5 + 3))

print(round(ans,2))
```

## Problem 2: Simpson's 1/3rd Rule

Write a python program to calculate the value of the integral ∫0 to π/4 (acos(x)+bsin(x))dx using Simpson's 1/3 rule. Divide the integration interval into N segments of equal width. The value of a, b and N will be specified as input. Print your answer rounded off up to two decimal places.
				
The input will consist of three lines, with the first and second lines specifying a and b respectively. The third line will contain the value of N.
					
Round the answer to two decimal places using the Python function round() as below: 
ans = round(ans, 2) 

Note: You should not use "scipy.integrate" package in your code.
					
Example:
INPUT:
2
3
100
OUTPUT:
2.29

```python
import numpy as np

a = float(input())
b = float(input())
N = int(input())

r = np.linspace(0,np.pi/4,N+1)

f0= a*np.cos(0) + b*np.sin(0)
fN= a*np.cos(np.pi/4) + b*np.sin(np.pi/4)

diff = r[1]-r[0]

ans = 0.0
ans+= (f0+fN)*(diff/3)

for i in range(1, len(r)-1):
    if i%2 != 0:
        ans+= ((4*diff)/3)*(a*np.cos(r[i]) + b*np.sin(r[i]))
    else:
        ans+= ((2*diff)/3)*(a*np.cos(r[i]) + b*np.sin(r[i]))

print(round(ans,2))
```

## Problem 3: 5 Point Legendre-Gauss Quadrature

Write a python program to calculate the approximate value of the integral ∫a to b exp(cx)cos(x)dx using five point Legendre-Gauss-quadrature rule. Take a, b and c as input.
			
The input will consist of three lines, where the first, second, and third lines are specifying a, b, and c respectively.

Note: Use the weights and the nodes provided in the textbook/notes for the five point legendre gauss quadrature

Round the answer to 2 decimal places using the python function round() as below: 
ans: round(ans,2) 
	
Example:
INPUT: 
  1
  2
 -2
	
OUTPUT:
0.01

```python
import numpy as np

weights = [0.23693, 0.47863, 0.56889, 0.47863, 0.23693]
roots = [-0.90618, -0.53847, 0.00000, 0.53847, 0.90618]

a = float(input())
b = float(input())
c = float(input())

ans = 0.0

for i in range(len(weights)):
    diff = (b-a)/2
    sums = (a+b)/2
    func = np.exp(c*(diff*roots[i] + sums))*np.cos(diff*roots[i] + sums)
    func*= weights[i]
    ans+= func

ans*= (b-a)/2

print(round(ans,2))
```
