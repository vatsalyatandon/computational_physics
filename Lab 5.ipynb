## Problem 1: First-Derivative

Write a python code to find the first derivative of the function f(x)=xlog10(x) at a given x0 for a given h(smallest division). First, use the two-point forward difference scheme and then two-point backward difference scheme for differentiation.

The input will be provided in two lines containing x0 and h respectively. Print the results obtained from both the schemes up to three decimal places as output in the two different lines. Also, print the absolute difference between the results (after rounding off) in next (third) line in the output.

Round the answer to three decimal places using the Python function round() as below: 
ans = round(ans, 3)

Note: You should not use any python library for the computation of derivative in your code.

[NOTE: You should maintain the following order of printing in your output -
OUTPUT:
result obtained from forward scheme 
result obtained from backward scheme
Absolute difference between them
]

Example:

INPUT:
1.0 
0.001 

OUTPUT:
0.435
0.434
0.001

```python
import numpy as np

x0 = float(input())
h = float(input())

def f(x):
    return x*np.log10(x)

forward_difference = (f(x0 + h) - f(x0))/h
backward_difference = (f(x0) - f(x0 - h))/h
print(round(forward_difference,3))
print(round(backward_difference,3))
print(round(abs(round(forward_difference,3) - round(backward_difference,3)),3))
```

## Problem 2: Third Derivative

Write a python code to find the third derivative of the function f(x)=sinxcosx using the second derivative of the function at a given x0 for a given h (smallest division).
	
The input will be provided in two lines containing x0 and h respectively. Print the result in the output after rounding off up to second decimal places.

Round the answer to two decimal places using the Python function round() as below:
ans = round(ans, 2)

Note: You should not use any python library for the computation of derivative in your code.

Hints: The third derivative can be obtained by successive calculation of second and first derivatives. This can be done by using 5 points centered at x0. First compute the second derivatives using three-point central difference scheme at 3 points centered at x0. Then use these values in the three-point central difference scheme for first derivative at x0, which will in turn give the third derivative at x0.

Example:

INPUT:
1.0
0.001
	
OUTPUT:
1.66

```python
import numpy as np

x0 = float(input())
h = float(input())

def f(x):
    return np.sin(x) * np.cos(x)

ans_plus_h = (f(x0) - 2*f(x0 + h) + f(x0 + 2*h))/(h**2)
ans_minus_h = (f(x0 - 2*h) - 2*f(x0 - h) + f(x0))/(h**2)

result = (ans_plus_h-ans_minus_h)/(2*h)
print(round(result,2))
```

## Problem 3: Euler Forward Scheme

A body of mass 2kg is attached to a spring with a spring constant 10. The differential equation governing the displacement of body y and time t is given by-
	
d2y/dt2 + 2dy/dt + 5y = 0

	
Find the displacement y(t=tfinal) using Euler-Forward scheme given that y(t=0) and dy/dt(t=0).
	
The input will consist of four lines, where the first three lines consist y(t=0), dy/dt(t=0), tfinal and followed by Δt in the last line. 

Round the answer to two decimal places using the Python function round() as below:
ans = round(ans, 2)


Example:
INPUT:
2.0 
-4.0 
1.5 
0.01 
	
OUTPUT:
-0.48

```python
y0 = float(input())
del_y0 = float(input())
t_final = float(input())
del_t = float(input())

def f(y,del_y):
    return -5*y + -2*del_y

y = y0
y_old=y0
del_y = del_y0
    
for i in range(int(t_final/del_t)):
    del_y+=f(y_old,del_y)*del_t
    y_old = y
    y+= del_y*del_t

print(round(y,2))
```

## Problem 4: Euler Forward Scheme

The differential equation which relates the current i, voltage V, resistance R and inductance L of a series electrical circuit, is given by -
	
	
didt+RLi=VL

	
Use Euler forward scheme to find the value of current i(t=tfinal) for a given initial condition i(t=0). Your code should take six inputs (each in a separate line), i(t=0), V,R,L, Δt and tfinal respectively. Round of your answer up to two decimal place.
	
Round the answer to two decimal places using the Python function round() as below:
ans = round(ans, 2)

Example:
	
INPUT:
	
0.0
240.0
20.0
50.0
0.01
2.0
	
OUTPUT:
6.62

```python
i0 = float(input())
v = float(input())
r = float(input())
l = float(input())
del_t = float(input())
t_final = float(input())

ans = i0
for i in range(int(t_final/del_t)):
    ans = ans + ((v/l) - (r/l)*(ans))*del_t

print(round(ans, 2))
```

## Problem 5: Euler Backward Scheme

Write a python code to solve the differential equation dx/dt+2x=0. Use Euler's backward scheme with time step, Δt to find the value of x(t) at given t=tfinal for a given initial condition x(t=0)=xinit. 
	
The input will be provided in three lines containing xinit, Δt and tfinal respectively. 
	
Round the answer to two decimal places using the Python function round() as below: 
ans = round(ans, 2)
	
	
Example:
INPUT:
1.0
0.01
2.0

OUTPUT:
0.02

```python
xinit = float(input())
del_t = float(input())
t_final = float(input())

ans = xinit

#Simplification:

#x(i+1) = x(i) + del_t * (f(x(i+1), t(i+1)))
#x(i+1) = x(i) + del_t * (-2 * x(i+1))
#x(i+1) = x(i)/(1 + 2*del_t)

for i in range(int(t_final/del_t)):
    ans = ans/(1 + 2*del_t)

print(round(ans,2))
```
